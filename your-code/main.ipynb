## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

```python
# YOUR CODE HERE
from fractions import Fraction

# Probability of taking a white ball
P_white = Fraction(10, 30)
# Probability of white first, then black (without replacement)
P_white_black_no_replace = Fraction(10, 30) * Fraction(8, 29)
# Probability of white first, then black (with replacement)
P_white_black_replace = Fraction(10, 30) * Fraction(8, 30)
# Probability of drawing 3 different colors
P_different_colors = Fraction(10, 30) * Fraction(12, 29) * Fraction(8, 28)
# Display results
P_white, P_white_black_no_replace, P_white_black_replace, P_different_colors
```

## Challenge 2

You want to go on picnic on Saturday. You hate rain, and would know to plan your weekend. To help you make a decision, you gather the following data about rainy days:

* 40% of all days are cloudy.
* About 50% of cloudy days are rainy. 
* About 20% of non-cloudy days are rainy.

What is the chance of rain during next Saturday?

```python
# YOUR CODE HERE
# Given probabilities
P_cloudy = 0.4
P_rain_given_cloudy = 0.5
P_rain_given_not_cloudy = 0.2

# P(Not Cloudy) = 1 - P(Cloudy)
P_not_cloudy = 1 - P_cloudy

# Law of Total Probability
P_rain = (P_rain_given_cloudy * P_cloudy) + (P_rain_given_not_cloudy * P_not_cloudy)

P_rain
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Given probabilities
P_A = 0.3  # Probability of A
P_B = 0.6  # Probability of B
P_A_and_B = 0.1  # Probability of A and B happening together

# Create a figure
fig, ax = plt.subplots()

# Create two circles for the Venn diagram
circle_A = plt.Circle((0.3, 0.5), 0.2, color='blue', alpha=0.5, label='A')  # Circle A
circle_B = plt.Circle((0.5, 0.5), 0.2, color='red', alpha=0.5, label='B')  # Circle B

# Add the circles to the plot
ax.add_artist(circle_A)
ax.add_artist(circle_B)

# Add the intersection part
# Intersection area: P(A  B) = 0.1
ax.text(0.4, 0.55, '0.1', color='black', fontsize=14, ha='center')

# Add text for the exclusive parts
ax.text(0.2, 0.5, f'{P_A - P_A_and_B:.2f}', color='black', fontsize=14, ha='center')  # P(A only)
ax.text(0.6, 0.5, f'{P_B - P_A_and_B:.2f}', color='black', fontsize=14, ha='center')  # P(B only)

# Set the limits of the plot
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')

# Remove axes and labels
ax.axis('off')

# Add title
plt.title("Venn Diagram for Events A and B", fontsize=16)

# Show the plot
plt.show()
```

## Challenge 4

 * If someone flies to China, the probability of getting Covid is 1%  
 * If someone flies to Italy, the probability of getting Covid is 20%  

I roll 1 dice with numbers 1, 2, 3, 4, 5, 6:  
 * If 1: I fly to Italy
 * If not 1: I fly to China

During my trip, I got Covid. What is the probability that I flew to China?

```python
# YOUR CODE HERE
# Given probabilities
P_Covid_given_C = 0.01  # Probability of getting Covid if flying to China
P_Covid_given_I = 0.20  # Probability of getting Covid if flying to Italy
P_C = 5 / 6  # Probability of flying to China
P_I = 1 / 6  # Probability of flying to Italy

# Total probability of getting Covid (P(Covid))
P_Covid = (P_Covid_given_C * P_C) + (P_Covid_given_I * P_I)

#Bayes' Theorem to find P(C | Covid)
P_C_given_Covid = (P_Covid_given_C * P_C) / P_Covid

# Print the result
print(f"P(C | Covid) = {P_C_given_Covid:.4f}")
```
